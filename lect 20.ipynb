```python
from skimage import io
from matplotlib import pyplot as plt
from skimage.transform import rescale, resize, downscale_local_mean
from skimage.filters import roberts, sobel, scharr, prewitt
from skimage.feature import canny
```

```python
img = io.imread("Images/a.jpg", as_gray = True)
rescaled_img = rescale(img,1.0/4.0, anti_aliasing = True)
resized_img = resize(img, (200,200))
downscale_img = downscale_local_mean(img,(4,3))
plt.subplot(1,3,1)
plt.imshow(rescaled_img)
plt.subplot(1,3,2)
plt.imshow(resized_img)
plt.subplot(1,3,3)
plt.imshow(downscale_img)
```

```python
edge_roberts = roberts(img)
plt.imshow(edge_roberts)
```

```python
edge_roberts = roberts(img)
edge_sobel = sobel(img)
edge_scharr = scharr(img)
edge_prewitt = prewitt(img)
plt.subplot(3,2,1)
plt.imshow(edge_roberts, cmap=plt.cm.gray)
plt.title('edge_roberts')
plt.subplot(3,2,2)
plt.imshow(edge_sobel, cmap=plt.cm.gray)
plt.title('edge_sobel')
plt.subplot(3,2,3)
plt.imshow(edge_scharr, cmap=plt.cm.gray)
plt.title('edge_scharr')
plt.subplot(3,2,4)
plt.imshow(edge_prewitt, cmap=plt.cm.gray)
plt.title('edge_prewitt')
plt.subplot(3,2,5)
plt.imshow(edge_roberts, cmap='gray')
plt.title('edge_roberts')
plt.subplot(3,2,6)
plt.imshow(img)
plt.title('img')
```
